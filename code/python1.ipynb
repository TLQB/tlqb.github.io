````python
class countdown(object):
  
  def __init__(self,start):
    self.start = start

  def __iter__(self):
    return countdown_iter(self.start)

class countdown_iter(object):

  def __init__(self, count):
    self.count = count

  def __next__(self):
    if self.count <= 0:
      raise StopIteration
    r = self.count
    self.count -= 1
    return r
````

````python
import time
from functools import wraps


def cached(timeout, logged=False):
    """Decorator to cache the result of a function call.
    Cache expires after timeout seconds.
    """
    def decorator(func):
        if logged:
            print ("-- Initializing cache for", func.__name__)
        cache = {}

        @wraps(func)
        def decorated_function(*args, **kwargs):
            if logged:
                print( "-- Called function", func.__name__)
            key = (args, frozenset(kwargs.items()))
            result = None
            if key in cache:
                if logged:
                    print( "-- Cache hit for", func.__name__, key
)
                (cache_hit, expiry) = cache[key]
                if time.time() - expiry < timeout:
                    result = cache_hit
                elif logged:
                    print ("-- Cache expired for", func.__name__, key)
            elif logged:
                print ("-- Cache miss for", func.__name__, key)

            # No cache hit, or expired
            if result is None:
                result = func(*args, **kwargs)

            cache[key] = (result, time.time())
            return result

        return decorated_function

    return decorator
````

````python
from inspect import Parameter, Signature

def make_args(names):
  return Signature(Parameter(name,Parameter.POSITIONAL_OR_KEYWORD) for name in names)

class Structure:
  _field = make_args([])
  def __init__(self, *args, **kwargs):
    bound = self._field.bind(*args,**kwargs)

    for name, value in bound.arguments.items():
      setattr(self,name,value)

class Person(Structure):
  _field = make_args(["name","age"])

  def _print(self):
    print(self.name,self.age)


class A(Structure):
  _field = make_args(["x","y","z","Z"])

  def _print(self):
    self.Z = self.x + self.y + self.z
    print(self.Z)
````

Đoạn code trên sử dụng mô hình hóa (metaprogramming) để xây dựng các lớp Structure và Person, A.

Trong đoạn code này, hàm make_args được sử dụng để tạo ra một Signature object từ một danh sách các tên tham số. Signature là một built-in class trong module inspect của Python, nó cho phép bạn xác định cấu trúc của một hàm hoặc phương thức dựa trên các tham số mà nó nhận.

Lớp Structure là một lớp cha, nó có một thuộc tính _field và một phương thức __init__. _field là một Signature object mà giá trị ban đầu là rỗng. Phương thức __init__ sử dụng bind method của Signature object để liên kết các tham số truyền vào với Signature object và đặt các giá trị tham số vào các thuộc tính của đối tượng Structure.

Lớp Person và A kế thừa từ lớp Structure và có một số thuộc tính riêng. Mỗi lớp có một _field khác nhau, được khởi tạo bằng hàm make_args và giá trị là một danh sách các tên tham số. Khi các lớp con này được khởi tạo, phương thức __init__ của lớp cha sẽ được gọi và sử dụng.

````python
import time 
from functools import wraps

def debug(prefix = ""):
    def decorator_class(cls,*args,**kwargs):
        def check(func):
            name_func = func.__name__
            @wraps(func)
            def check_time(*args, **kwargs):
                print(prefix,"\nStart run function {}".format(name_func))
                start_time = time.time()
                result = func(*args, **kwargs)
                print("End run function {}, time = {} \n".format(name_func,time.time() - start_time))
                return result
            return check_time
        methods = vars(cls)
        for name,method in methods.items():
            if callable(method):
                setattr(cls,name,check(method))

        return cls  
    return decorator_class


class Structure:
    _field = []
    def __init__(self, *args):
        if len(args) != len(self._field):
            print(len(args))
            raise TypeError("ERROR")

        for name, value in zip(self._field,args):
            setattr(self,name,value)


class ValidateObject:

    def __init__(self, name_obj = None,obj_type = None, min_value = 0, min_length = 0, gender_value = ["male","female"]):
        self.name_obj = name_obj
        self.obj_type = obj_type
        self.min_value = min_value
        self.min_length = min_length
        self.gender_value = gender_value

    def __set__(self,instance,value):
        if not isinstance(value, self.obj_type):
            raise TypeError("{} EXPECTED {}".format(self.name_obj,self.obj_type))

        if isinstance(value,int) and value < self.min_value:
            raise AttributeError("{} EXPECTED VALUE > {}".format(self.name_obj,self.min_value))

        if isinstance(value,str) and len(value) < self.min_length:
            raise AttributeError("{} EXPECTED LENGTH > {}".format(self.name_obj,self.min_length))

        if self.name_obj == "gender" and value not in self.gender_value:
            raise AttributeError("{} EXPECTED IN {}".format(self.name_obj,self.gender_value))

        # super().__set__(instance,value)
        instance.__dict__[self.name_obj] = value


@debug(prefix="***Debug***")
class Person(Structure):

    _field = ["name","age","gender"]

    #@markdown ## Input value field "name" :
    obj_type = str #@param ["int", "str", "float"] {type:"raw"}
    min_length = 10 #@param {type:"number"}

    name = ValidateObject(name_obj ="name",obj_type = obj_type, min_length = min_length)

    #@markdown ## Input value field "age":
    
    obj_type = str #@param ["int", "str", "float"] {type:"raw"}
    min_value = 0 #@param {type:"number"}
    age = ValidateObject(name_obj ="age",obj_type = obj_type, min_value = min_value)

    #@markdown ## Input value field "gender":
    obj_type = str  #@param ["int", "str","float"]  {type:"raw"}
    gender = ValidateObject(name_obj ="gender",obj_type = obj_type)

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self,value):
        if not isinstance(value,int):
            raise TypeError("ERROR")
        self._age = value 

    def study(self, course_name : str ):
        print("{} - {}".format(self.name, course_name))

    @classmethod
    def params_strings(cls,params : str):

        strings = params.split(",")
        name,age,gender = strings 

        return cls(name, int(age),gender) 

#@markdown ## Input value param:
name_param = "hello girl ..." #@param {type:"string"}
age_param = 10 #@param {type:"number"}
gender_param = "female" #@param ["male", "female"]
p = Person(name_param,age_param,gender_param)
````

````python
def debug_update(prefix = "Debug :"):
  def debug_all_in_one(cls):
    def debug(func):
      func_name = func.__qualname__
      @wraps(func)
      def print_msg(*agrs,**kwagrs):
        print(prefix,func_name)
        return func(*agrs,**kwagrs)
      return print_msg

    methods = vars(cls)
    for name, method in methods.items():
      if callable(method):
        setattr(cls,name,debug(method))
    return cls

  return debug_all_in_one
````
